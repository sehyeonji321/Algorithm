```python
# 18870 -> 인덱스를 전부 순환하며 조사 -> 시간초과

N = int(input())
x = list(map(int, input().split()))

for idx in range(N):
  cnt = 0
  is_append = []
  for j in range(N):
    if x[j] < x[idx] and x[j] not in is_append:
      cnt += 1
      is_append.append(x[j])
  print(cnt, end=" ")
```

```python
# 18870
# X를 정렬하며 중복 제거 -> 인덱스가 X'의 값 -> 근데 이것도 시간초과

N = int(input())
x = list(map(int, input().split()))
x_ = list(set(x))
x_.sort()

for i in range(N):
  print(x_.index(x[i]), end=" ") # index 함수가 bottleneck
```

```python
# 18870

N = int(input())
x = list(map(int, input().split()))
x_ = list(set(x))
x_.sort()

val_idx = {val: idx for idx, val in enumerate(x_)}
for i in range(N):
  print(val_idx[x[i]], end=" ")
```

```python
# 1620 -> 시간초과
N, M = map(int, input().split())
dic = []

for i in range(N):
  name = input()
  dic.append(name)

for j in range(M):
  query = input()

  try:
    query = int(query)
    print(dic[query-1])
  except:
    q_idx = {val:idx for idx, val in enumerate(dic)}
    print(q_idx[query]+1)
```

```python
# 1620

N, M = map(int, input().split())
name_idx = {}
idx_name = {}

for i in range(N):
  idx, name = i+1, input()
  name_idx[name] = idx
  idx_name[idx] = name

for j in range(M):
  query = input()
  try:
    query = int(query)
    print(idx_name[query])

  except:
    print(name_idx[query])
```

```python
# 1764

non_seen = set()
non_heard = set()

N, M = map(int, input().split())

for i in range(N):
  non_seen.add(input())

for j in range(M):
  non_heard.add(input())

names = non_seen & non_heard
names = sorted(names)
print(len(names))
print(*names, sep="\n")
```
