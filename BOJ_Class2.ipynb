```python
# 30802
N = int(input())
size = list(map(int, input().split()))
q = []
T, P = map(int, input().split())

for k in size:
  if k % T == 0:
    q.append(k // T)
  else:
    q.append((k // T)+1)


print(sum(q))
print(N // P, N % P)
```

```python
# 2798

N, M = map(int, input().split())
numbers = list(map(int, input().split()))
sum = []

for i in range(len(numbers)-2):
  for j in range(i+1, len(numbers)-1):
    for k in range(j+1, len(numbers)):
      sum.append(numbers[i]+numbers[j]+numbers[k])

sum.sort(reverse = True)

for num in sum:
  if num <= M:
    print(num)
    break
```

```python
# 1018

def howmuch(board, m, n): # 보드검사, 탐색횟수 m*n
  cnt_b = []
  cnt_w = []
  current_b = 0
  current_w = 0
  for i in range(m):
    for j in range(n):
      for a in range(8):
        for b in range(8):
          if a % 2 == 0:
            if b % 2 == 0 and board[i+a][j+b] == "W":
              current_b += 1
            elif b % 2 == 1 and board[i+a][j+b] == "B":
              current_b += 1
            elif b % 2 == 0 and board[i+a][j+b] == "B":
              current_w += 1
            elif b % 2 == 1 and board[i+a][j+b] == "W":
              current_w += 1
          else:
            if b % 2 == 0 and board[i+a][j+b] == "B":
              current_b += 1
            elif b % 2 == 1 and board[i+a][j+b] == "W":
              current_b += 1
            elif b % 2 == 0 and board[i+a][j+b] == "W":
              current_w += 1
            elif b % 2 == 1 and board[i+a][j+b] == "B":
              current_w += 1

      cnt_b.append(current_b)
      cnt_w.append(current_w)
      current_b = 0
      current_w = 0

  result = min(min(cnt_b), min(cnt_w))
  return result

h, w = map(int, input().split())
board = []

for i in range(h):
  board.append(list(input())) # 이제, board는 list형태의 각 row를 담은 list

m = h - 7
n = w - 7

print(howmuch(board, m, n))
```

```python
# 2775
people = [[]]
for i in range(1,15): # 14칸의 base 배열
  people[0].append(i)

for i in range(1, 15):
  people.append([]) # i번째 층을 만들고
  for j in range(14): # j+1호의 인원은
    people[i].append(0)
    for k in range(j+1): # 이전 층의 1호부터 j+1호까지의 합
      people[i][j] += people[i-1][k]

T = int(input())
for t in range(T):
  k = int(input())
  n = int(input())

  print(people[k][n-1])
```
