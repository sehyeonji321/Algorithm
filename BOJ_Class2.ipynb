```python
# 1018

def howmuch(board, m, n): # 보드검사, 탐색횟수 m*n
  cnt_b = []
  cnt_w = []
  current_b = 0
  current_w = 0
  for i in range(m):
    for j in range(n):
      for a in range(8):
        for b in range(8):
          if a % 2 == 0:
            if b % 2 == 0 and board[i+a][j+b] == "W":
              current_b += 1
            elif b % 2 == 1 and board[i+a][j+b] == "B":
              current_b += 1
            elif b % 2 == 0 and board[i+a][j+b] == "B":
              current_w += 1
            elif b % 2 == 1 and board[i+a][j+b] == "W":
              current_w += 1
          else:
            if b % 2 == 0 and board[i+a][j+b] == "B":
              current_b += 1
            elif b % 2 == 1 and board[i+a][j+b] == "W":
              current_b += 1
            elif b % 2 == 0 and board[i+a][j+b] == "W":
              current_w += 1
            elif b % 2 == 1 and board[i+a][j+b] == "B":
              current_w += 1

      cnt_b.append(current_b)
      cnt_w.append(current_w)
      current_b = 0
      current_w = 0

  result = min(min(cnt_b), min(cnt_w))
  return result

h, w = map(int, input().split())
board = []

for i in range(h):
  board.append(list(input())) # 이제, board는 list형태의 각 row를 담은 list

m = h - 7
n = w - 7

print(howmuch(board, m, n))
```

```python
# 2775
people = [[]]
for i in range(1,15): # 14칸의 base 배열
  people[0].append(i)

for i in range(1, 15):
  people.append([]) # i번째 층을 만들고
  for j in range(14): # j+1호의 인원은
    people[i].append(0)
    for k in range(j+1): # 이전 층의 1호부터 j+1호까지의 합
      people[i][j] += people[i-1][k]

T = int(input())
for t in range(T):
  k = int(input())
  n = int(input())

  print(people[k][n-1])
```

```python
# 2750 -> 중복이 없는 경우는 돼야하는거 아닌가? 왜 안되는거지

def mini(seq, start):
  M = seq[start]

  for i in range(start+1, len(seq)):
    if seq[i] < M:
      M = seq[i]

  return M

N = int(input())
seq = []
for i in range(N):
  seq.append(int(input()))

# 선택 정렬
for k in range(len(seq)):
  m = mini(seq, k)
  seq[k], seq[seq.index(m)] = seq[seq.index(m)], seq[k]

print(seq)
```

```python
# 2750 -> class2는 아니지만 재밌어보여서 풀어봤다.

def mini_idx(seq, start):
  min_val = seq[start]
  min_idx = start

  for i in range(start+1, len(seq)):
    if min_val > seq[i]:
      min_val = seq[i]
      min_idx = i

  return min_idx

N = int(input())
seq = [int(input()) for _ in range(N)]

# 선택 정렬
for k in range(len(seq)):
  idx = mini_idx(seq, k)
  seq[k], seq[idx] = seq[idx], seq[k]
  # seq[k], seq[mini_idx(seq, k)] = seq[mini_idx(seq, k)], seq[k] 이렇게하면 안된다.

for i in range(len(seq)): print(seq[i])

```

```python
# 2751 삽입정렬로 풀이할 시 시간초과!

N = int(input())
seq = [int(input()) for i in range(N)]

for idx in range(0, N-1):
  tmp = seq[idx+1]
  for j in range(idx, -1, -1):
    if tmp < seq[j]:
      seq[j+1] = seq[j]
      seq[j] = tmp
    else:
      seq[j+1] = tmp
      break

for i in range(N):
  print(seq[i])
```
