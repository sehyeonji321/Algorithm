```python
# 30802
N = int(input())
size = list(map(int, input().split()))
q = []
T, P = map(int, input().split())

for k in size:
  if k % T == 0:
    q.append(k // T)
  else:
    q.append((k // T)+1)


print(sum(q))
print(N // P, N % P)
```

```python
# 2231

N = int(input())

if N == 1:
  print(0)

for i in range(1,N):
  gen = i + sum(map(int, str(i)))

  if gen == N:
    print(i)
    break
  elif i == N-1:
    print(0)
```

```python
# 2798

N, M = map(int, input().split())
numbers = list(map(int, input().split()))
sum = []

for i in range(len(numbers)-2):
  for j in range(i+1, len(numbers)-1):
    for k in range(j+1, len(numbers)):
      sum.append(numbers[i]+numbers[j]+numbers[k])

sum.sort(reverse = True)

for num in sum:
  if num <= M:
    print(num)
    break
```

```python
# 15829

def hashfunction(seq):
  result = 0
  for i in range(len(seq)):
    result += (seq[i]-96)*(31**i)
  result = result % 1234567891
  return result


_ = int(input())
seq = list(map(ord, input()))
H = hashfunction(seq)
print(H)
```

```python
# 1259

while True:
  n = input()
  if int(n) == 0: break
  p = list(map(int, n))

  if p == p[::-1]:
    print("yes")
  else:
    print("no")
```
