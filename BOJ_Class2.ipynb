```python
# 30802
N = int(input())
size = list(map(int, input().split()))
q = []
T, P = map(int, input().split())

for k in size:
  if k % T == 0:
    q.append(k // T)
  else:
    q.append((k // T)+1)


print(sum(q))
print(N // P, N % P)
```

```python
# 2231

N = int(input())

if N == 1:
  print(0)

for i in range(1,N):
  gen = i + sum(map(int, str(i)))

  if gen == N:
    print(i)
    break
  elif i == N-1:
    print(0)
```
